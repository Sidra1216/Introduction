## 2. Show 2D Histogram of Image

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# 画像の読み込み
img = cv2.imread('model/saba_qamar_01.jfif', 0)

# ヒストグラムの取得
img_hist_cv = cv2.calcHist([img], [0], None, [256], [0, 256])

# ヒストグラムの表示
plt.plot(img_hist_cv)
plt.show()
```

## 3. Get 2D Histogram RGB(maximum value) of Multiple Images

```python
#OpenCVでRGBのヒストグラムを取得

def RGB_img(path):
    
    import numpy as np
    import matplotlib.pyplot as plt
    import cv2
    from skimage import io

    img = cv2.imread(path)
    b, g, r = img[:,:,0], img[:,:,1], img[:,:,2]

    hist_b = cv2.calcHist([b],[0],None,[256],[0,256])
    hist_g = cv2.calcHist([g],[0],None,[256],[0,256])
    hist_r = cv2.calcHist([r],[0],None,[256],[0,256])

    plt.plot(hist_r, color='r', label="r")
    plt.plot(hist_g, color='g', label="g")
    plt.plot(hist_b, color='b', label="b")
    plt.legend()
    plt.show() 
    
    return hist_r,hist_g, hist_b
```

```python
# read multiple images

import glob
image_path_list = glob.glob("model/*")
```

```python
# You can use the #enumerate function to retrieve the element index and the element at the same time.

import matplotlib.pyplot as plt
from skimage import io

for i, j in enumerate(image_path_list):
    plt.title('Model_'+ str(i+1))
    plt.imshow(io.imread(j))
    plt.show()
    r,g,b = RGB_img(j)
    print("r",max(r[0:250]),"g",max(g[0:250]),"b",max(g[0:250]))
```
